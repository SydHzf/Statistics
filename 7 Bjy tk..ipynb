```python
import pandas as pd
import numpy as np
```

```python
# importing file 
# we need Sheet = "Creditors Oct21 v051121" 

dataset = pd.read_excel("Aged payable schedule - October 2021.XLSX" , sheet_name = "Creditors Oct21 v051121" , header = None)
```

```python
# seeing our Data

dataset.head()
```

```python
# kaam ki rows and columns nikaal rhe

data = dataset.iloc[: , 3:21]
data
```

```python
columns = data.iloc[8]
columns
```

```python
# setting columns of our df

data.columns = columns
```

```python
# seeing our data

data.head(10)
```

```python
# is column se "Name" , "Vendor" , Account Nikaal lo

dataset[0]
```

```python
# getting Names :

names_list = []

for index , row in dataset.iterrows():

    name_head = row[0]
    name = row[4]

    if name_head == " Name" :
        names_list.append(name)

names_list
```

```python
# getting "Vendors"

vendors_list = []

for index , row in dataset.iterrows():

    vendor_head = row[0]
    vendor = row[4]

    if vendor_head == " Vendor" :
        vendors_list.append(vendor)

vendors_list
```

```python
# getting "Accounts"

accounts_list = []

for index , row in dataset.iterrows():

    account_head = row[1]
    account = row[2]

    if account_head == "**" :
        accounts_list.append(account)

accounts_list
```

```python
# removing null valies

data = data[data["Doc. Date"].notnull()]
data.head(5)
```

```python
# Removing first row 

data = data.iloc[1: , :]
data.head(10)
```

```python
# inserting our columns

data.insert(0 , "Account" , np.nan)
data.insert(1 , "Vendor" , np.nan)
data.insert(2 , "Name" , np.nan)
```

```python
# checking if inserted

data.head(5)
```

```python
# now adding value one by one :

for index , row in data.iterrows():

    inner_col = row["DocumentNo"]
    
    if len(accounts_list) == 0:
        break
    
    data["Account"][index] = accounts_list[0]
    data["Vendor"][index] = vendors_list[0]
    data["Name"][index] = names_list[0]
    
    if inner_col == "DocumentNo":
        del names_list[0]
        del vendors_list[0]
        del accounts_list[0]
```
