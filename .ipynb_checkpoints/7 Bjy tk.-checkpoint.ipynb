```python
# importing libraries

import pandas as pd

import numpy as np

pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
# importing file 
# we need Sheet = "Creditors Oct21 v051121" 

age_original = pd.read_excel("Aged payable schedule - October 2021.XLSX" , sheet_name = "Creditors Oct21 v051121" , header = None)
```

```python
# seeing our Data

age_original.head()
```

```python
# kaam ki rows and columns nikaal rhe

age = age_original.iloc[: , 3:21]
age
```

```python
#  kaam ke columns 

columns = age.iloc[8]
columns
```

```python
# setting columns of our df

age.columns = columns
```

```python
# seeing our data

age.head(10)
```

```python
# is column se "Name" , "Vendor" , Account Nikaal lo

age_original[0]
```

```python
# getting Names :

names_list = []

for index , row in age_original.iterrows():

    name_head = row[0]
    name = row[4]

    if name_head == " Name" :
        names_list.append(name)

names_list
```

```python
# getting "Vendors"

vendors_list = []

for index , row in age_original.iterrows():

    vendor_head = row[0]
    vendor = row[4]

    if vendor_head == " Vendor" :
        vendors_list.append(vendor)
   
vendors_list
```

```python
# getting "Accounts"

accounts_list = []

for index , row in age_original.iterrows():

    account_head = row[1]
    account = row[2]

    if account_head == "**" :
        accounts_list.append(account)
   
accounts_list
```

```python
# removing null valies

age = age[age["Doc. Date"].notnull()]
age.head(5)
```

```python
# Removing first row 

age = age.iloc[1: , :]
age.head(10)
```

```python
# inserting our columns

age.insert(0 , "Account" , np.nan)
age.insert(1 , "Vendor" , np.nan)
age.insert(2 , "Name" , np.nan)
```

```python
# checking if inserted

age.head(5)
```

```python
# now adding value one by one :

for index , row in age.iterrows():

    inner_col = row["DocumentNo"]
    
    if len(accounts_list) == 0:
        break
    
    age["Account"][index] = accounts_list[0]
    age["Vendor"][index] = vendors_list[0]
    age["Name"][index] = names_list[0]
    
    if inner_col == "DocumentNo":
        
        del names_list[0]
        del vendors_list[0]
        del accounts_list[0]
```

```python
# seeing our data

age.head(15)
```

```python
# now removing sub columns :

age = age[age["   Amount in DC"] != "   Amount in DC"]
age.head(10)
```

#### Pivot :

```python
# changing our "   Amount in DC" column into float

age["   Amount in DC"] = age["   Amount in DC"].astype(float)
```

```python
age.pivot_table(index = "Curr",\
                values = "   Amount in DC" , aggfunc = sum)                   
```

```python
# getting the columns we need (pehle is liye nhi kia danial bhai ne end mn banaya tha)

desired_columns = [0,1,2,3,4,5,6,7,10,11,13,14,16]
```

```python
age = age.iloc[: , desired_columns]
```
