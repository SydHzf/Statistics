#  Cumulative Balance (LC) of all Assets

```python
data1 = data.iloc[:,[0,1,13]]
data1.head()
```

```python
data1 = data1[data1["Posting Date"].notnull() == True]
data1
```

```python
data1["Posting Date"].unique()
```

```python
posting_type = data1["Posting Date"][data1["Posting Date"].str.contains(r"[a-z]+")  == True].unique()
posting_type
```

```python
balance = {}
temp = ""
for i in np.array(data1):
    pos_date = i[0]
    cb = i[2]
    
    if pos_date in posting_type:
        temp = pos_date
    
    
    elif temp not in balance:
        balance[temp] = cb
        
        
    elif temp in balance:
        balance[temp] += cb
        
        
balance
```

```python
summary1 = pd.DataFrame(list(balance.values()),index = list(balance.keys()), columns = ["Cumulative Balance (LC)"])
summary1.index.name = "Type"

summary1
```

```python
total = summary1["Cumulative Balance (LC)"].sum()
balance["All"] = total
balance
```

```python
summary1 = pd.DataFrame(list(balance.values()),index = list(balance.keys()), columns = ["Cumulative Balance (LC)"])
summary1.index.name = "Type"

print("Task 1:  Comulative Balance of all types of Posting Date ")
summary1
```

```python
np.arange(summary1.shape[0]-1)
```

```python
summary1.index[:-1]
```

```python
plt.style.use('dark_background')

plt.figure(figsize=(16,7))
plt.bar(x = summary1.index[:-1], height=summary1["Cumulative Balance (LC)"][:-1], color = "blue")
plt.title("Cummulative Balance of All Types",{'fontsize' :"30"} )

plt.xticks(summary1.index[:-1],rotation = 90)
#.ticklabel_format(axis = 'y')
plt.show()
```

# Cumulative Balance By Head

```python
concat = []
for i in np.array(data1):
    pos = i[0]
    due = i[1]

    concat.append(str(pos) +" "+str(due) )

concat
```

```python
data1.insert(2,"Head",concat)
data1.head()
```

```python
hn = data1["Head"][data1["Head"].str.contains(r"[a-z]+")]
hn
```

```python
arr = []
temp = ""
for i in data1["Head"]:
    
    if i in np.array(hn):
         temp = i
    
    arr.append(temp)

data1.insert(3,"Head Name",arr)
data1
```

# Cumulative Balance By Months

```python
date = "01/06/2019"
```

```python
pattern = "%m/%d/%Y"
t = dt.datetime.strptime(date,pattern)
t.strftime("%B")
```

```python
Date = []
pattern = "%d/%m/%Y"

for i in data1["Posting Date"]:
    
    if i in posting_type:
        Date.append("")
        
    else:
        d = dt.datetime.strptime(str(i),pattern)
        m = d.strftime("%B")
        Date.append(m)
        
Date
```

```python
data1.insert(0,"Month",Date)
data1.tail(10)
```

```python
data1["Month"].unique()
```

```python
group_month = data1.groupby("Month")
months = group_month.sum()
months
```

```python
month = \
data1.pivot_table(index="Month", values = "Cumulative Balance (LC)", aggfunc="sum",margins=True)

month
```

```python
new_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December','All']
summary3 = month.reindex(new_order, axis=0)

print("Task 3: Cummulative Balance By Month")
summary3
```
